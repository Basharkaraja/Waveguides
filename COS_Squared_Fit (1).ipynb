```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit



# Define the model function
def cos_squared_model(x, A , J , B):
    return  A * np.cos(J * x )**2  +B

# Fit the data
initial_guess = [0.1,.005,1]  # Initial guess for the parameters
params, covariance = curve_fit(cos_squared_model, x_data, y, p0=initial_guess)

# Extract fitted parameters
A_fit,J_fit,B_fit = params

# Generate fitted data
y_fit = cos_squared_model(x_data, A_fit,J_fit,B_fit)

# Plot the data and the fit
plt.figure(figsize=(10, 6))
plt.plot(x_data, y, 'bo', label='Data')
plt.plot(x_data, y_fit, 'r-', label='Cos^2 Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Cos^2 Fit to Data for d = 3.5')
plt.show()

# Output the fitted parameters
params
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the exponential decay model function
def exp_decay_model(x, A, J, B):
    return A * np.exp(-J * x) + B

# Fit the data
initial_guess = [1, 0.001, 1]  # Initial guess for the parameters
params, covariance = curve_fit(exp_decay_model, X_j, d, p0=initial_guess)

# Extract fitted parameters
A_fit, J_fit, B_fit = params

# Generate fitted data
y_fit = exp_decay_model(X_j, A_fit, J_fit, B_fit)

# Plot the data and the fit
plt.figure(figsize=(10, 6))
plt.plot( d,X_j, 'bo', label='Data')
plt.plot( y_fit,X_j, 'r-', label='Exponential Decay Fit')
plt.xlabel('Distance d (m)')
plt.ylabel('Coupling coefficinet J (1/m)')
plt.legend()
plt.title('Coupling Coefficinet J vs Distance d')
plt.show()

# Output the fitted parameters
params
```
